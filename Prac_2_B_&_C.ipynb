2 primary ways of handling missing values:
*   Deleting the Missing values
*   Imputing the Missing Values



```python
#deleteing the missing values(deleting entire row)
df1=df.dropna(axis=0)
print(df1.isnull().sum())
```

```python
#deleting entire column
df2=df.drop(['Customer_Segment'],axis=1)
print(df2.isnull().sum())
```

```python
#Imputing the Missing Value
#replace missing values with 0
df1=pd.read_csv('/content/Sales_with_NaNs_v1.3.csv')
df1["Customer_Segment"] =  df1["Customer_Segment"].fillna(0)
print(df1["Customer_Segment"].isnull().sum())
```

```python
#replacing missing values with mean
df2=pd.read_csv('/content/Sales_with_NaNs_v1.3.csv')
df2["Sales_Before"] =  df2["Sales_Before"].fillna(df2["Sales_Before"].mean())
print(df2["Sales_Before"].isnull().sum())
```

```python
#replacing with mode
df3=pd.read_csv('/content/Sales_with_NaNs_v1.3.csv')
df3["Customer_Satisfaction_After"] =  df3["Customer_Satisfaction_After"].fillna(df3["Customer_Satisfaction_After"].mode()[0])
print(df3["Customer_Satisfaction_After"].isnull().sum())
```

```python
#replacing with median
df4=pd.read_csv('/content/Sales_with_NaNs_v1.3.csv')
df4["Age"] =  df4["Age"].fillna(df4["Age"].median())
print(df4["Age"].isnull().sum())
```

```python
# Impute Missing Values Using Sci-kit Learn Librar
# Univariate Approach
import numpy as np
from sklearn.impute import SimpleImputer
df4=pd.read_csv('/content/Titanic-Dataset.csv')
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
df4['Age'] = imp.fit_transform(df4[['Age']])
print(imp.transform(df4[['Age']].head(6)))
```

```python
df5=pd.read_csv('/content/Titanic-Dataset.csv')
num_cols = df5.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df5.select_dtypes(include=['object']).columns

#converting text to numbers
from sklearn.preprocessing import LabelEncoder
le_dict = {}
for col in cat_cols:
    le = LabelEncoder()
    df5[col] = df5[col].astype(str)
    df5[col] = le.fit_transform(df4[col])
    le_dict[col] = le

# Multivariate Approach
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
impute_it = IterativeImputer()
df5_imputed = impute_it.fit_transform(df5)
df5_imputed = pd.DataFrame(df5_imputed, columns=df5.columns)
print(impute_it.transform(df5_imputed.head(6)))
```

```python
#(KNNImputer)
from sklearn.impute import KNNImputer
impute_knn = KNNImputer(n_neighbors=2)
impute_knn.fit_transform(df5)
print(impute_knn.transform(df5.head(6)))
```
